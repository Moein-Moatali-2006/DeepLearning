```python
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from tensorflow.keras import models,layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
dataset_path = "/content/drive/MyDrive/7-7-Faces"

img_dt_generator = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.1,
)

dataset_train = img_dt_generator.flow_from_directory(
    dataset_path,
    subset="training",
    shuffle=True,
    target_size=(224,224)
)

dataset_validation = img_dt_generator.flow_from_directory(
    dataset_path,
    subset="validation",
    target_size=(224,224),
    shuffle=False,
)
```

```python
model = tf.keras.applications.VGG16(include_top=False,weights="imagenet",input_shape=(224,224,3))
```

```python
my_model = models.Sequential([
    model,
    layers.Flatten(),
    layers.Dense(2000,activation="relu"),
    layers.Dense(1000,activation="relu"),
    layers.Dense(14,activation="softmax")
])
```

```python
my_model.summary()
```

```python
my_model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
history = my_model.fit(dataset_train,epochs=20,validation_data=dataset_validation)
```
