```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from insightface.app import FaceAnalysis
```

```python
# app = FaceAnalysis(providers=['CUDAexcutionProvider'])
app = FaceAnalysis(name="buffalo_s",providers=['CUDAexcutionProvider'])
app.prepare(ctx_id=0,det_size=(640,640))
```

```python
image_1 = cv2.imread("image1.jpg")
image_1 = cv2.cvtColor(image_1,cv2.COLOR_BGR2RGB)
plt.imshow(image_1)
plt.show()
```

```python
result = app.get(image_1)
```

```python
embedding_1 = result[0]["embedding"]
```

```python
image_2 = cv2.imread("image2.jpg")
image_2 = cv2.cvtColor(image_2,cv2.COLOR_BGR2RGB)
plt.imshow(image_2)
plt.show()
```

```python
result_2 = app.get(image_2)
```

```python
embedding_2 = result_2[0]["embedding"]
```

```python
np.sqrt(np.sum((embedding_1 - embedding_2)**2))
```
