```python
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.cifar10

(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python

model = tf.keras.models.Sequential([
  # Deep Learning
                        # 28 * 28
  tf.keras.layers.Conv2D(32,(3,3),activation="relu",input_shape=(32,32,3)), # 26 *26
  tf.keras.layers.Conv2D(40,(5,5),activation="relu"), # 22 * 22
  tf.keras.layers.Conv2D(25,(3,3),activation="relu"), # 20 * 20
  tf.keras.layers.Conv2D(10,(5,5),activation="relu"), # 16 * 16
  tf.keras.layers.MaxPooling2D(), # 8 * 8
  tf.keras.layers.Conv2D(32,(3,3),activation="relu"), # 6 * 6
  tf.keras.layers.Conv2D(32,(3,3),activation="relu"), # 4 * 4
  tf.keras.layers.MaxPooling2D(), # 2 * 2
  tf.keras.layers.Flatten(), # 1 * 4


  # Machine lesrning
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation='softmax')
])
```

```python

model.compile(optimizer='adam',
  loss='sparse_categorical_crossentropy',
  metrics=['accuracy'])

model.fit(x_train, y_train, epochs=5)
model.evaluate(x_test, y_test)
```
