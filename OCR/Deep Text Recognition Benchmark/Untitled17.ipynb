```python
import os
import zipfile

# Define dataset directory
dataset_dir = '/content/Dataset'

# Create Dataset folder if not exist
if not os.path.exists(dataset_dir):
    os.makedirs(dataset_dir)

# Unzip dataset files if needed (skip if manually uploaded)

# Prepare train and validation folders
train_dir = os.path.join(dataset_dir, 'train')
validation_dir = os.path.join(dataset_dir, 'validation')

if not os.path.exists(train_dir):
    os.makedirs(train_dir)
if not os.path.exists(validation_dir):
    os.makedirs(validation_dir)
```

```python
import xml.etree.ElementTree as ET

# Train GT
gt_train_path = os.path.join(dataset_dir, 'gt_train.txt')
lines = []

for filename in os.listdir(train_dir):
    if filename.endswith('.xml'):
        xml_path = os.path.join(train_dir, filename)
        tree = ET.parse(xml_path)
        root = tree.getroot()
        label = root.find('object').find('name').text
        img_filename = filename.replace('.xml', '.jpg')
        img_path = os.path.join('data', img_filename)
        lines.append(f"{img_path}\t{label}\n")

with open(gt_train_path, 'w') as f:
    f.writelines(lines)
```

```python
# Validation GT
gt_val_path = os.path.join(dataset_dir, 'gt_validation.txt')
lines = []

for filename in os.listdir(validation_dir):
    if filename.endswith('.xml'):
        xml_path = os.path.join(validation_dir, filename)
        tree = ET.parse(xml_path)
        root = tree.getroot()
        label = root.find('object').find('name').text
        img_filename = filename.replace('.xml', '.jpg')
        img_path = os.path.join('data', img_filename)
        lines.append(f"{img_path}\t{label}\n")

with open(gt_val_path, 'w') as f:
    f.writelines(lines)
```

```python
import shutil

# Create data/train and data/validation folders
if not os.path.exists('data/train'):
    os.makedirs('data/train')
if not os.path.exists('data/validation'):
    os.makedirs('data/validation')

# Copy train images
for filename in os.listdir(train_dir):
    if filename.endswith('.jpg'):
        shutil.copy(os.path.join(train_dir, filename), 'data/train/')

# Copy validation images
for filename in os.listdir(validation_dir):
    if filename.endswith('.jpg'):
        shutil.copy(os.path.join(validation_dir, filename), 'data/validation/')
```
